```python
with open("output.txt", "w") as f:
    f.write("param money := " + str(money) + ";")
    f.write("\n")
    """
    Other Paramters go here if needed
    """
    f.write("set States := " )
    f.write("\n")
    for i in df['State FIPS Code'].value_counts().index:
        f.write(str(i))
        f.write("\n")
    f.write(";")
    f.write("\n")
    
    f.write("param entry_cost:=")
    f.write("\n")
    for i in df['State FIPS Code'].value_counts().index:
        f.write(str(i)+ "     " + str(np.random.randint(5,20) * 100_000))
        f.write("\n")
    f.write(";")
    f.write("\n")
    
    f.write("set Counties := " )
    f.write("\n")
    for i in df.index:
        f.write(str(i))
        f.write("\n")
    f.write(";")
    f.write("\n")
    
    for feature in rv:
        f.write("param " + feature.replace(" ","").replace("-","").replace("%","").replace("/","") + ":=")
        for i in df.index:
            f.write(str(i) + "        " + str(df[feature][i]))
            f.write("\n")
        f.write(";")
        f.write("\n")
    
    f.write("param Pairs: ")
    for j in df['State FIPS Code'].value_counts().index:
        f.write(str(j) + " ")
    f.write(":=")
    f.write("\n")
    for i in df.index:
        f.write(str(i) + "    ")
        for j in df['State FIPS Code'].value_counts().index:
            if(df['State FIPS Code'][i] == j):
                f.write('1    ')
            else:
                f.write('0    ')
        f.write("\n")
    f.write(";")

print("Done!")
```
